## **Problem Statement**

Write a function that takes a list of numbers and returns the sum of those numbers.

```python
def add_many_number(numbers:list[int])->int:
    total : int = 0

    for number in numbers:
        total += number

    return total

def main():
    list_of_numbers : list[int] = [1, 2, 3, 4, 5]

    sum_of_numbers : int = add_many_number(list_of_numbers)

    print(f"List of Numbers: {list_of_numbers}")
    print(f"Total Sum of Numbers: {sum_of_numbers}")

if __name__ == '__main__':
    main()
```
