```python
def main():
    kilos_of_mass : float = float(input("Enter kilos of mass: "))
    C : int = 299792458
    energy_in_joules : float = kilos_of_mass * (C ** 2)

    print("e = m * C^2...")
    print(f"m = {kilos_of_mass} kg")
    print(f"C = {C} m/s")
    print(f"{energy_in_joules} joules of energy!")

if __name__ == '__main__':
    main()
```
