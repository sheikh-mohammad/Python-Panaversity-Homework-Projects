## **Problem Statement**

Simulate rolling two dice, three times. Prints the results of each die roll. This program is used to show how variable scope works.

```python
# Random module is imported to generate random numbers
import random # Importing the random module

# Number of sides on each die
num_sides = 6 # Standard six-sided dice

# Function to roll two dice and display the result
def roll_dice():
    # Simulate rolling two dice
    dice_1 = random.randint(1, num_sides) # Roll first die
    dice_2 = random.randint(1, num_sides) # Roll second die
    # Display the result
    print(f"You rolled a {dice_1} and a {dice_2} (Total: {dice_1 + dice_2})")

# Main function to execute the dice rolls
def main():
    # Roll the dice three times
    roll_dice() # First roll
    roll_dice() # Second roll
    roll_dice() # Third roll
    
# Entry point of the script
if __name__ == "__main__": # Check if the script is being run directly
    main() # Call the main function to start the program
```
