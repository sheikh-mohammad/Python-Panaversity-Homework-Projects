## **Problem Statement**

Simulate rolling two dice, and prints results of each roll as well as the total.

```python
import random

NUM_OF_SIDES : int = 6

def main():
    roll_dice_1 : int = random.randint(1, NUM_OF_SIDES)
    roll_dice_2 : int = random.randint(1, NUM_OF_SIDES)

    total : int = roll_dice_1 + roll_dice_2

    print(f"The result of dice 1 is {roll_dice_1} and of dice 2 is {roll_dice_2}. (Total of roll of dices is: {total})")

if __name__ == '__main__':
    main()
```
